```python
import pandas as pd
import numpy as ny
```

```python
#Lê dataframe criado com histórico de emprestimos
emprestimos = pd.read_csv('/Users/Clarice Nascimento/0. Alura/emprestimos.csv')
```

```python
#Importando base de dados com as informações dos alunos contendo numero de matricula e curso por ano
matricula_alunos_2010 = pd.read_excel('/Users/Clarice Nascimento/0. Alura/d6/matricula_alunos.xlsx',sheet_name='Até 2010',skiprows=1)
matricula_alunos_2011 = pd.read_excel('/Users/Clarice Nascimento/0. Alura/d6/matricula_alunos.xlsx',sheet_name='Após 2010',skiprows=1)
matricula_alunos_2010 = matricula_alunos_2010.rename(columns={'Número da matrícula': 'Matrícula','Tipo de vinculo':'Vinculo','Curso':'Curso'})
matricula_alunos_2011 = matricula_alunos_2011.rename(columns={'Matricula ': 'Matrícula','Tipo de vinculo':'Vinculo','Cursos':'Curso'})
```

```python
#Agregando as importações em um unico dataframe
matricula_alunos = pd.concat([matricula_alunos_2010,matricula_alunos_2011], ignore_index=True)
matricula_alunos
```

```python
#importando arquivo em formato json
cadastro_alunos = pd.read_json('/Users/Clarice Nascimento/0. Alura/d6/cadastro_alunos.json')
cadastro_alunos
```

```python
#Realizando a transposição das informações de interesse contidas no arquivo de json em formato de dataframe 
cadastro_alunos = pd.read_json(cadastro_alunos.registros[0])
cadastro_alunos
```

```python
#Renomeando colunas do dataframe extraido do arquivo json
cadastro_alunos = cadastro_alunos.rename(columns={'matricula_ou_siape': 'Matrícula','tipo_vinculo_usuario':'Vinculo','curso':'Curso'})
cadastro_alunos
```

```python
#Concatenando os dataframes obtidos com a importação de arquivos em excel e json para um unico dataframe
info_alunos = pd.concat([matricula_alunos,cadastro_alunos], ignore_index=True)
info_alunos
```

```python
#Filtrando somente alunos de gradução para analise exploratoria
alunos_graduacao = emprestimos.query('tipo_vinculo_usuario == "ALUNO DE GRADUAÇÃO"')
#Incluindo o curso dos alunos no dataframe de observação
alunos_graduacao = alunos_graduacao.merge(info_alunos,  how='left', left_on='matricula_ou_siape', right_on='Matrícula')
alunos_graduacao = alunos_graduacao.drop(columns=['Matrícula','Vinculo'])
#Filtrando somente registros a partir do ano de 2015 e cursos selecionados para analise
alunos_graduacao['data_emprestimo'] = pd.to_datetime(alunos_graduacao['data_emprestimo']) 
alunos_graduacao = alunos_graduacao.query('data_emprestimo > 2015')
alunos_graduacao = alunos_graduacao.query("Curso == ['BIBLIOTECONOMIA','CIÊNCIAS SOCIAIS','COMUNICAÇÃO SOCIAL','DIREITO','FILOSOFIA','PEDAGOGIA']")
alunos_graduacao
```

```python
#Extraindo e criando coluna de ano a partir da data de emprestimo
alunos_graduacao['ano'] = alunos_graduacao['data_emprestimo'].dt.year
```

```python
#Criando um novo dataframe contendo a quantidade de empretimos realizadas por ano em cada curso de graduação
total_curso_ano = alunos_graduacao.groupby(['ano','Curso'])['data_emprestimo'].count().reset_index(name='qtds')
total_curso_ano
```

```python
#Criando a tabela pivot formato 1
pivot_table = total_curso_ano.pivot_table(index='ano', columns='Curso', values='qtds', aggfunc='sum',margins=True, margins_name='Total')
pivot_table
```

```python
#Criando a tabela pivot formato 2
pivot_table = total_curso_ano.pivot_table(index='Curso', columns='ano', values='qtds', aggfunc='sum',margins=True, margins_name='Total')
pivot_table
```
