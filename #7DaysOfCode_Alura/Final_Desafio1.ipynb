```python
import pandas as pd
import numpy as ny
import glob
import pyarrow.parquet as pq
```

```python
#Lista todos os arquivos CSV no diretório
lista_arquivos = glob.glob('/Users/Clarice Nascimento/0. Alura/d1/emprestimos-*.csv')
```

```python
#Inicializa um DataFrame vazio para armazenar os dados combinados
base_emprestimos = pd.DataFrame()
```

```python
#Loop sobre cada arquivo CSV e concatena ao DataFrame combinado
for file in lista_arquivos:
    df = pd.read_csv(file)
    base_emprestimos = pd.concat([base_emprestimos, df])
```

```python
#Salva o DataFrame combinado em um arquivo CSV
base_emprestimos.to_csv('/Users/Clarice Nascimento/0. Alura/banco_emprestimos.csv', index=False)
base_emprestimos.head()
```

```python
#Contagem de repetidos
base_emprestimos.value_counts()
```

```python
#Apaga repetidos e conta novamente
base_emprestimos = base_emprestimos.drop_duplicates()
base_emprestimos.value_counts()
```

```python
#Lê arquivo parquet e salva em um dataframe
dados_exemplares = pq.read_table('/Users/Clarice Nascimento/0. Alura/d1/dados_exemplares.parquet').to_pandas()
dados_exemplares
# ou pd.read_parquet
```

```python
#Mescla as duas bases de dados a partir da coluna de codigo de barras
emprestimos = pd.merge(base_emprestimos,dados_exemplares, on = 'codigo_barras',how='inner')
emprestimos.to_csv('/Users/Clarice Nascimento/0. Alura/emprestimos.csv', index=False)
emprestimos
```
