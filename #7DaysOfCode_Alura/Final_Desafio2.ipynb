```python
import pandas as pd
import numpy as ny
import glob
import pyarrow.parquet as pq
```

```python
#Cria dataframe com classificação universal de temas
df_cdu = {'Faixa de Código': ['000 a 099', '100 a 199', '200 a 299', '300 a 399', '400 a 499', '500 a 599', '600 a 699', '700 a 799', '800 a 899', '900 a 999'],
    'Descrição': ['Generalidades, Ciência e conhecimento', 'Filosofia e psicologia', 'Religião', 'Ciências sociais', 'Classe vaga: Provisoriamente não ocupada', 'Matemática e ciências naturais', 'Ciências aplicadas', 'Belas artes', 'Linguagem, Língua, Linguística', 'Geografia, Biografia, História']}
cdu = pd.DataFrame(df_cdu)
cdu
```

```python
#Lê dataframe criado com histórico de emprestimos
emprestimos = pd.read_csv('/Users/Clarice Nascimento/0. Alura/emprestimos.csv')
emprestimos.head()
```

```python
#Função para obter a descrição com base no número de localização
def obter_descricao(localizacao):
    for faixa, descricao in zip(cdu['Faixa de Código'], cdu['Descrição']):
        inicio, fim = map(int, faixa.split(' a '))
        if inicio <= localizacao <= fim:
            return descricao
    return None
```

```python
#Adiciona uma nova coluna 'Descrição' ao dataframe 'emprestimos' com base no número de localização
#Método apply() do Pandas para aplicar essa função a cada valor na coluna 'Localização' do DataFrame
#Função obter_descricao() para mapear cada número de localização para a descrição correspondente das faixas de códigos.
emprestimos['descricao'] = emprestimos['localizacao'].apply(obter_descricao)
#Salva o dataframe atualizado de volta para o arquivo CSV de empréstimos
emprestimos.to_csv('/Users/Clarice Nascimento/0. Alura/emprestimos.csv', index=False)
emprestimos.head()
```
